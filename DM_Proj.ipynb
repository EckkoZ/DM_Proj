### 2.Preprocessing Job Descriptions (NLTK)

```python
import html
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Updated Preprocessing Function
def preprocess_text(text):
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    
    # HTML Decode
    text = html.unescape(text)

    # Remove non-alphabetic words (e.g., `nbsp`, `rsquo`)
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Tokenize words
    words = word_tokenize(text.lower())
    words = [word for word in words if word.isalnum()]  # Remove punctuation
    words = [word for word in words if word not in stop_words]  # Remove stopwords
    words = [lemmatizer.lemmatize(word) for word in words]  # Lemmatization
    
    return " ".join(words)
```

```python
# Apply preprocessing to Job Descriptions

from tqdm import tqdm

tqdm.pandas() 
df_jobs['Cleaned_Description'] = df_jobs['Job.Description'].astype(str).progress_apply(preprocess_text)
```

### 3.TFIDF

```python
# Convert text to numerical features using TF-IDF
vectorizer = TfidfVectorizer(max_features=500)  # Limit features for efficiency
X = vectorizer.fit_transform(df_jobs['Cleaned_Description'])
```

### 4. Elbow method

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Find the optimal number of clusters using the Elbow Method
wcss = []
K_range = range(1, 15)  # Testing clusters from 1 to 14

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)  # Store within-cluster sum of squares (WCSS)

# Plot the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(K_range, wcss, marker='o', linestyle='--')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
plt.title("Elbow Method for Optimal k")
plt.show()
```

### 5.Apply K-Means Clustering and cisualize

```python
num_clusters = 5  # Define number of clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df_jobs['Cluster'] = kmeans.fit_predict(X)

# Reduce dimensionality for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray())

# Visualize Clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df_jobs['Cluster'], palette="viridis", alpha=0.7)
plt.title("Job Descriptions Clustering (PCA Projection)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Cluster")
plt.show()

# Display Sample Results
print(df_jobs[['Job.ID', 'Title', 'Cluster']].head(10))
```

```python
### Check Clusters and Outliers
print(df_jobs['DBSCAN_Cluster'].value_counts())
```

```python
### Interpret DBSCAN Clusters

from collections import Counter

def get_common_words_per_dbscan_cluster(df, n_words=15):
    cluster_words = {}
    for cluster in df['DBSCAN_Cluster'].unique():
        all_words = " ".join(df[df['DBSCAN_Cluster'] == cluster]['Cleaned_Description']).split()
        word_freq = Counter(all_words).most_common(n_words)
        cluster_words[cluster] = [word for word, freq in word_freq]
    
    return cluster_words

dbscan_common_words = get_common_words_per_dbscan_cluster(df_jobs)
```

```python
#### Generate Word Clouds for DBSCAN Clusters
from wordcloud import WordCloud

fig, axes = plt.subplots(1, len(dbscan_common_words), figsize=(15, 5))

for i, (cluster, words) in enumerate(sorted(dbscan_common_words.items())):
    wordcloud = WordCloud(width=500, height=500, background_color='white').generate(" ".join(words))
    axes[i].imshow(wordcloud, interpolation='bilinear')
    axes[i].set_title(f"DBSCAN Cluster {cluster}")
    axes[i].axis("off")

plt.show()
```

- DBSCAN might not be the best choice for clustering Job Descriptions (JD) because it struggles with high-dimensional data, such as TF-IDF text representations, where distance-based density estimation becomes less meaningful. Additionally, choosing an appropriate eps (neighborhood radius) is challenging, as textual data often has varying similarity levels—too small an eps results in excessive noise, while too large merges everything into one cluster. Moreover, job descriptions exhibit uneven density across industries (e.g., tech vs. food service), and DBSCAN performs best when clusters have uniform density, making it less effective for this use case.

```python
import numpy as np
from scipy.sparse import issparse

# Ensure X_dense is defined
if 'X' in locals():  
    if issparse(X):  
        X_dense = X.toarray()  # Convert sparse to dense
    else:
        X_dense = X  
else:
    raise ValueError("Variable X is not defined. Make sure your dataset is loaded.")

print("Shape of X_dense:", X_dense.shape)  # Should print (samples, features)
```

```python
from tqdm import tqdm
from kmodes.kmodes import KModes

best_k = 10  # Adjust based on dataset size
print(f"Fitting K-Modes with k={best_k}...")

# Use tqdm to track iterations
kmodes = KModes(n_clusters=best_k, init='Huang', n_init=5, verbose=1)

# ✅ Wrap the fit_predict process in tqdm
with tqdm(total=kmodes.max_iter, desc="K-Modes Progress", unit="iteration") as pbar:
    cluster_labels = kmodes.fit_predict(X_dense)
    pbar.update(kmodes.max_iter)  # Manually update progress

print("K-Modes Clustering complete!")
```
